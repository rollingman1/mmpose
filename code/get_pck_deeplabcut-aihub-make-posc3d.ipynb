```python
def transform_csv(keypoint_path):
    keypoint_filename = keypoint_path.split("/")[-1]
    print(keypoint_filename)
    keypoint_df = pd.read_csv(keypoint_path, header=None)
    
    colnames = [str(keypoint_df.iloc[1, i]) + '_' + str(keypoint_df.iloc[2, i]) for i in range(len(keypoint_df.columns))]
    colnames[0] = "sequence"
    keypoint_df.columns = colnames
    keypoint_df = keypoint_df.iloc[3:, :].reset_index(drop=True)
    
    keypoint_df.iloc[:, 1:] = keypoint_df.iloc[:, 1:].apply(pd.to_numeric)
    keypoint_df = keypoint_df[[i for i in keypoint_df.columns if (('_x' in i)|('_y' in i)|('sequence' in i)|('threshold_' in i))]]
    return keypoint_df
```

```python
def get_pck(target_df, predict_df, dataframe, name):
    col = []
    
    if "threshold_nan" in target_df.columns: # threshold = area
        threshold = target_df.threshold_nan
        target_df = target_df.drop(["threshold_nan"], axis=1)
    else: # threshold = 양쪽 귀 거리의 1/3
        threshold = ((target_df.right_ear_start_x-target_df.left_ear_start_x).pow(2) +
                 (target_df.right_ear_start_y-target_df.left_ear_start_y).pow(2)).pow(1/2) * (1/3)
    print(target_df.shape, predict_df.shape)
    col.append(min(target_df.shape[0], predict_df.shape[0]))
    
    
   
    # 행이름(파일명) 동일한지 체크
    if target_df.shape[0] != predict_df.shape[0]:
        predict_df = predict_df.iloc[np.where(predict_df.sequence.isin(target_df.sequence))[0],:].reset_index(drop=True)    

#         predict_seq=predict_df.loc[:,'sequence']
#         target_seq=target_df.loc[:,'sequence']

#         target_seq = target_seq.sort_index(ascending=False)
#         predict_seq = predict_seq.sort_index(ascending=False)

#         for i in range(predict_df.shape[0]):
#             strings = target_seq[i] .split('/')
#             if strings[-1] != predict_seq[i]:
#                 target_df = target_df.drop(index=i, axis=0)
#                 i-=1
#         print( target_df.shape[0])
#         print( predict_df.shape[0])
        


    # x와 y의 제곱차 더해주고 루트
    diff2 = target_df.iloc[:,1:].sub(predict_df.iloc[:,1:]).pow(2)
    diff2_x = diff2[[i for i in diff2.columns if ('_x' in i)]]
    diff2_x.columns = [i[:-2] for i in diff2_x.columns]
    diff2_y = diff2[[i for i in diff2.columns if ('_y' in i)]]
    diff2_y.columns = [i[:-2] for i in diff2_y.columns]
    diff_df = diff2_x.add(diff2_y).pow(1/2)
    
    # target 없는 열 제거 및 threshold 없는 행 제거
    diff_df = diff_df.iloc[:,np.where(diff_df.isnull().sum(0)!=diff_df.shape[0])[0]]
    diff_df = diff_df[threshold.notnull()]
    threshold = threshold[threshold.notnull()]
    
    # keypoint 별 정확도
    correct_num = diff_df.apply(lambda x: x < threshold).sum(axis=0)
    keypoint_num = diff_df.notnull().sum(axis=0)
    accuracy = (correct_num / keypoint_num * 100).sort_values(ascending=False)
    keypoint_accuracy_df = pd.DataFrame({"accuracy":accuracy, "correct_num":correct_num, "keypoint_num":keypoint_num}).sort_values(["accuracy"], ascending=False)
    print("\nkeypoint 별 정확도")
    # print(accuracy)
    print("accuracy_mean:", keypoint_accuracy_df.accuracy.mean())
    
    # 이미지 별 정확도
    correct_num = diff_df.apply(lambda x: x < threshold).sum(axis=1)
    keypoint_num = diff_df.notnull().sum(axis=1)
    accuracy = correct_num / keypoint_num * 100
    image_accuracy_df = pd.DataFrame({"accuracy":accuracy, "correct_num":correct_num, "keypoint_num":keypoint_num}).sort_values(["accuracy"], ascending=False)
    print("\n이미지 별 정확도")
    print("accuracy_mean:", image_accuracy_df.accuracy.mean())
    
    # 총 정확도
    correct_num = diff_df.apply(lambda x: x < threshold).sum().sum()
    keypoint_num = diff_df.notnull().sum().sum()
    pck = correct_num / keypoint_num * 100
    print("\n최종결과")
    print("testset:", diff_df.shape[0])
    print("correct/keypoint:", correct_num, "/", keypoint_num)
    print("pck:", pck)
    col.append(diff_df.shape[0])
    col.append(pck)
    col.append(keypoint_accuracy_df.accuracy.mean())
    col.append(image_accuracy_df.accuracy.mean())
    dataframe[name] = col
    
    return pck
```

```python
from skimage import io

def MakeLabeledPlots(folder, DataCombined, cfg, Labels, Colorscheme, cc, scale):
    tmpfolder = str(folder) + "_labeled"
    auxiliaryfunctions.attempttomakefolder(tmpfolder)
    for index, imagename in enumerate(DataCombined.index.values):
        image = io.imread(os.path.join("/home/petpeotalk/DeepLabCut-core/Test-Suhyun-2021-05-21/labeled-data/user_keypoint_out_prev",
                                       imagename))
        plt.axis("off")

        if np.ndim(image) == 2:
            h, w = np.shape(image)
        else:
            h, w, nc = np.shape(image)

        plt.figure(
            frameon=False, figsize=(w * 1.0 / 100 * scale, h * 1.0 / 100 * scale)
        )
        plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

        plt.imshow(image, "gray")
        if index == 0:
            print("They are stored in the followiFng folder: %s." % tmpfolder)  # folder)
        print(DataCombined)

        for c, bp in enumerate(cfg["bodyparts"]):
            plt.plot(
                DataCombined["DLC_resnet152_TrainTestJul2shuffle1_1030000"][bp]["x"].values[index],
                DataCombined["DLC_resnet152_TrainTestJul2shuffle1_1030000"][bp]["y"].values[index],
                Labels[cc],
                color=Colorscheme(c),
                alpha=cfg["alphavalue"],
                ms=7,
            )

        plt.xlim(0, w)
        plt.ylim(0, h)
        plt.axis("off")
        plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
        plt.gca().invert_yaxis()

        # plt.savefig(str(Path(tmpfolder)/imagename.split(os.sep)[-1]))
        plt.savefig(
            os.path.join(tmpfolder, str(imagename))
        )  # create file name also on Windows for Unix projects (and vice versa)
        plt.close("all")
```
